# The documents were devided into following time bins
and some of the few top words were extracted to see how weigtage of each word has changed

<h3> ['1951-60', '1961-70', '1971-80', '1981-90', '1991-2000', '2001-2010', '2011-18']</h3>

```python
from pprint import pprint
#get words you want to compare
timelist_runway=np.zeros(7)
for i in range(7):
    freq = df_topic0_time[i]['P_Score'][df_topic0_time[i]['Top_words']=='runway']
    freq=freq.values.tolist()
    if(len(freq)>0):
        timelist_runway[i]=freq[0]
    else:
        timelist_runway[i]=0
print("timelist_runway",timelist_runway)

# word2
timelist_surface=np.zeros(7)
for i in range(7):
    freq = df_topic0_time[i]['P_Score'][df_topic0_time[i]['Top_words']=='surface']
    freq=freq.values.tolist()
    if(len(freq)>0):
        timelist_surface[i]=freq[0]
    else:
        timelist_surface[i]=0
print("timelist_surface",timelist_surface)

# word3
timelist_airport=np.zeros(7)
for i in range(7):
    freq = df_topic0_time[i]['P_Score'][df_topic0_time[i]['Top_words']=='airport']
    freq=freq.values.tolist()
    if(len(freq)>0):
        timelist_airport[i]=freq[0]
    else:
        timelist_airport[i]=0
print("timelist_airport",timelist_airport)

# word4
timelist_strip=np.zeros(7)
for i in range(7):
    freq = df_topic0_time[i]['P_Score'][df_topic0_time[i]['Top_words']=='strip']
    freq=freq.values.tolist()
    if(len(freq)>0):
        timelist_strip[i]=freq[0]
    else:
        timelist_strip[i]=0
print("timelist_strip",timelist_strip)

# word5
timelist_approach=np.zeros(7)
for i in range(7):
    freq = df_topic0_time[i]['P_Score'][df_topic0_time[i]['Top_words']=='approach']
    freq=freq.values.tolist()
    if(len(freq)>0):
        timelist_approach[i]=freq[0]
    else:
        timelist_approach[i]=0
print("timelist_approach",timelist_approach)
```

```python
#plot graph
import matplotlib.pyplot as  plt
fig = plt.figure(figsize=(20,8))
ax = plt.subplot(111)
X=[0,1,2,3,4,5,6]
plt.plot(X,timelist_approach, 'r', label='approach')
plt.plot(X,timelist_runway, 'b', label='runway')
plt.plot(X,timelist_airport, 'g', label='airport')
plt.plot(X,timelist_surface, 'c', label='surface')
plt.plot(X,timelist_strip, 'k', label='strip')

x_label=['1951-60','1961-70','1971-80','1981-90','1991-2000','2001-2010','2011-18']
plt.xticks(X,label=x_label)
ax.set_xticklabels(x_label, fontsize=15)

plt.xlabel("time slice years", fontsize=20)
plt.ylabel("p score for word", fontsize=20)
plt.title("Word probabilities for a topic at different time intervals", fontsize=20)

plt.legend(loc='lower left', fontsize = 'x-large')
plt.grid()
plt.show()
```
