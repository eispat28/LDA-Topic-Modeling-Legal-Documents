# 2. DTM for 10 year time slices for 5 topics

```python
# load the model and reduce the size further

dictionary = corpora.Dictionary(data_clean)
dictionary.filter_extremes(no_below=15, no_above=0.70)
print('Number of unique tokens: %d' % len(dictionary))

corpus = [dictionary.doc2bow(doc) for doc in data_clean]
print("Corpus is ready")
```

```python
# get the time slices
import re
f = open('yearList.csv', 'r')
strings = re.findall(r"[0-9]+", f.read())
strings

years = np.array(strings,dtype=int)
years

np.array_split(years, 10)

# group of 10 years from 1951-1960, 1961-1970,..
#upperbound =[1960,1970,1980,1990,]
timeslice =[6,10,57,313,474,492,401]

a =years[years>=2011]
b= a[a<2011]
len(a)
sum(timeslice)
```

```python
import time
start = time.time()
timeslice10years =[6,10,57,313,474,492,401]
ldaSeqModel_10years_range = ldaseqmodel.LdaSeqModel(corpus=corpus, id2word=dictionary, time_slice=timeslice10years, num_topics=5)
end = time.time()
print(end - start)
```

```python
ldaSeqModel_10years_range.save("ldaSeqModel_10years_range")
```

```python
list(zip(np.arange(0,6,1),ldaSeqModel_10years_range.print_topics(time=0)))
```

```python
ldaSeqModel_10years_range.print_topic_times(topic=1) 
```

# 3. DTM for 10 years with 25 topics

```python
%%time
# load LDA model
#saving the model
from gensim.test.utils import datapath

# Load the pretrained model from disk.
temp_file = datapath("lda_model_25")
lda_model_25_topics = gensim.models.ldamodel.LdaModel.load(temp_file)

import time
start = time.time()
timeslice10years =[6,10,57,313,474,492,401]
ldaSeqModel_10years_range_25_topics = ldaseqmodel.LdaSeqModel(corpus=corpus, id2word=dictionary, time_slice=timeslice10years, num_topics=25, initialize='gensim',lda_model=lda_model_25_topics, chain_variance=0.05)
ldaSeqModel_10years_range_25_topics.save("ldaSeqModel_10years_range_25_topics")
end = time.time()
print(end - start)
print("Completed")
```

```python
ldaSeqModel_10years_range_25_topics.print_topic_times(topic=0)
```
